```python
imagem = cv2.resize(imagem, (800, 600))
```

```python
imagem.shape
```

```python
imagem_cinza = plt.imshow(cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY))
plt.axis('off')
plt.show('Imagem Cinza')
```

```python

detector_facial = cv2.CascadeClassifier('/Users/emersonmendes/Python/Vis√£o Computacional Guia Completo/Cascades/haarcascade_frontalface_default.xml')
```

```python
deteccoes = detector_facial.detectMultiScale(imagem)
```

```python
deteccoes
```

```python
len(deteccoes)
```

```python
for x, y, w, h in deteccoes:
    cv2.rectangle(imagem, (x, y), (x+w, y+h), (0, 255, 255), 5)
plt.imshow(cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
imagem = cv2.imread('/Users/emersonmendes/Python/Vis√£o Computacional Guia Completo/Images/people2.jpg')
imagem = cv2.resize(imagem, (800, 600))
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
deteccoes = detector_facial.detectMultiScale(imagem_cinza, scaleFactor=1.25)
for x, y, w, h in deteccoes:
    cv2.rectangle(imagem, (x, y), (x+w, y+h), (0, 255, 0), 5)
plt.imshow(cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
